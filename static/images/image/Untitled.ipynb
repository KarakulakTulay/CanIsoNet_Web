```python
import pandas as pd
```

```python
ENST_list_df = pd.read_csv('../ENST_list_to_download.tsv')
ENST_list_list = list(ENST_list_df.iloc[:,0])
ENST_list_list
```

```python
#!/usr/bin/env python
# coding: utf-8

import requests
from bs4 import BeautifulSoup

#Set URL of transcript image only for Homo_sapiens
URL = "https://grch37.ensembl.org/Homo_sapiens/Component/Transcript/Summary/image?db=core;t="
#List 


for ENST in ENST_list_list:

        #Get image src URL
    soup = BeautifulSoup(requests.get(URL+ENST).content, "html.parser")
    imgURL = soup.find("img").attrs.get("src")

        #Write image
    response = requests.get(f"https://grch37.ensembl.org{imgURL}")

    file = open(f"{ENST}.png", "wb")
    file.write(response.content)
    file.close()
```
